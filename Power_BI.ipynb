```python
import pandas as pd
from PIL import Image
```

```python
df = pd.read_csv("IMDB_Movies.csv")
pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
df.head()
```

```python
df.isnull().sum()
```

##### Fill the numeric null values with 0. This will reduce null values in this dataset.

```python
df[['num_critic_for_reviews','director_facebook_likes',
    'actor_3_facebook_likes','actor_1_facebook_likes',
    'num_voted_users','cast_total_facebook_likes',
    'facenumber_in_poster','num_user_for_reviews',
    'actor_2_facebook_likes','movie_facebook_likes']] = df[['num_critic_for_reviews','director_facebook_likes',
                                                            'actor_3_facebook_likes','actor_1_facebook_likes',
                                                            'num_voted_users','cast_total_facebook_likes',
                                                            'facenumber_in_poster','num_user_for_reviews',
                                                            'actor_2_facebook_likes','movie_facebook_likes']].fillna(0)
```

```python
df.isnull().sum()
```

##### Remove all null values in this data set

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.dtypes
```

##### In the above data, some data types are mismatched, so change all data types correctly.

```python
df['num_critic_for_reviews'] = df['num_critic_for_reviews'].astype(int)
df['duration'] = df['duration'].astype(int)
df['director_facebook_likes'] = df['director_facebook_likes'].astype(int)
df['actor_3_facebook_likes'] = df['actor_3_facebook_likes'].astype(int)
df['actor_1_facebook_likes'] = df['actor_1_facebook_likes'].astype(int)
df['facenumber_in_poster'] = df['facenumber_in_poster'].astype(int)
df['num_user_for_reviews'] = df['num_user_for_reviews'].astype(int)
df['title_year'] = df['title_year'].astype(int)
df['actor_2_facebook_likes'] = df['actor_2_facebook_likes'].astype(int)
```

Drop duplicates if any

```python
df.shape
```

```python
df = df.drop_duplicates('movie_imdb_link').reset_index(drop=True)
```

```python
df.shape
```

```python
df.head(2)
```

```python
df['language'].value_counts()
```

##### In the above data, we found that the language column has one misleading value (None). Remove that value and keep the remaining rows.

```python
df = df[df['language'] != 'None']
```

```python
df.head()
```

```python
df.to_csv('IMDB_movies.csv', index=False)
```

# Data Analytics

## Movie Genre Analysis

```python
Dashboard_1 = Image.open('Dashboard 1.jpg')
Dashboard_1 = Dashboard_1.resize((1500,700))
Dashboard_1
```

*1. In this dataset, we have a total of 3,660 movies, and there are 22 unique genres.*<br>
*2. The mean and median IMDb ratings for documentary genres are higher, but the variance is significant, so we cannot reach a single conclusion.*<br>
*3. Genres like history, biography, war, drama, animation, and film-noir (old black and white films) are generally well-liked by people, and the IMDb ratings are higher for these genres.*<br>
*4. Genres like horror, comedy, action, family, and sci-fi are not as popular. The mean and median ratings for these genres are low, and the variance is also high.*<br>
*5. We cannot be very confident about genres like documentary, music, family, sci-fi, and fantasy due to the high variance in their ratings.*

## Language Analysis

```python
Dashboard_2 = Image.open('Dashboard 2.jpg')
Dashboard_2 = Dashboard_2.resize((1500,700))
Dashboard_2
```

## Movie Duration Analysis

```python
Dashboard_3 = Image.open('Dashboard 3.jpg')
Dashboard_3 = Dashboard_3.resize((1500,700))
Dashboard_3
```

*1. The duration of the movies ranges from 37 minutes to 330 minutes.*<br>
*2. Movies with a duration of less than 125 minutes tend to have lower IMDb ratings, while movies longer than 125 minutes generally have an IMDb rating above 6.*<br>
*3. Analyzing the minimum IMDb ratings shows that movies with a duration below approximately 125 minutes have lower ratings, while those above approximately 125 minutes tend to have higher ratings.*<br>
*4. Movies longer than 275 minutes are not very stable in terms of ratings, with more drastic deviations.*<br>
*5. For an average good IMDb rating, the ideal duration is between 130 to 225 minutes.*<br>

## Director Analysis
